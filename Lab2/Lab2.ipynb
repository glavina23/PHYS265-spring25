## Part 1: The Ideal Case

Calculate time for test mass to reach bottom (Use 0 drag, alpha = 0)

$$
\frac{d^{2}y}{dt^2} = -g + {\alpha} |\frac{dy}{dt}|^{\gamma}
$$

$$t = \text{time}$$
$$y = \text{height}$$
$$g = \text{gravitational acceleration at surface}$$
$$\alpha = \text{drag coefficient}$$
$$\gamma = \text{speed dependence of the drag}$$

### Calculation Using Simple Freefall equations, no drag

$$
y = \frac{1}{2}at^2
$$

$$
t = \sqrt{\frac{2y}{a}}
$$

$$
\text{Assume y = -4000 m,}\,\,\,\,    \text{a = -9.8 m/s}
$$

```python
# (1)
    # y = (0.5)*a*t**2
    # t = sqrt(2y/a)
    # y = -4e3 m, a = -9.8 = -g0

y = -4e3 # meters
a = -g0
t_simple = np.sqrt(2*y/a)
print(f"In simple freefall kinematics, the time to fall 4km from the surface of Earth is {t_simple} seconds")
```

### Calculation using $\frac{d^{2}y}{dt^2} = -g + {\alpha} |\frac{dy}{dt}|^{\gamma}$, no drag

```python
# (2)
    # d2ydt2 = -g - a(dydt)**gamma
    # let dydt = v
    # dvdt = -g - a*v**gamma
    # s = (y, v)

# Function
def dvdt(t, s):
    y, v = s
    dydt = v 
    dvdt = -g - np.abs(a*v**gamma)
    return [dydt, dvdt]

# Constants 
a = 0
g = g0 # not variable until next part
gamma = 2 # not used until next part, though

# Time span
t0, tf = 0, 30
t_eval = np.linspace(t0, tf, 100)

# Initial conditions
y0 = [0, 0] # initial position and velocity

soln = solve_ivp(fun = dvdt, t_span=(t0, tf), y0=y0, t_eval = t_eval)

print(soln)

# Plot
t = soln.t
y = soln.y[0]
v = soln.y[1]

fig, ax = plt.subplots()
ax.plot(t, y, color = "royalblue")
ax.set_ylabel("Position (m)", color = "royalblue")
ax.tick_params(axis="y", labelcolor="royalblue")

ax.axhline(y=-4000, color = "black", linestyle = "--", linewidth = 0.8, label = "Bottom of the Shaft (m)")

ax.set_xlabel("Time (s)")
ax.legend()

ax2 = ax.twinx()
ax2.plot(t, v, color="deeppink", linestyle = ":")
ax2.set_ylabel("Velocity (m/s)", color = "deeppink")
ax2.tick_params(axis="y", labelcolor="deeppink")


fig.suptitle("Figure 1")
ax.set_title("Position and Velocity of Test Mass Falling Down Shaft")

fig.tight_layout();


# Get Event
def crash(t, s):
    return s[0]+4000

soln_crash = solve_ivp(fun = dvdt, t_span=(t0, tf), y0=y0, t_eval = t_eval, events = crash)
#print(soln_crash)

#print(soln_crash.t_events[0])

print(f"In the case of 0 drag and constant g, the time to hit the bottom of the shaft is {soln_crash.t_events[0]} seconds")
print(f"Analyic time - Numeric time = {t_simple - soln_crash.t_events[0]}")
```

## Part 2: Including Drag and a Variable $g$

$$
g(r) = {g_0}(\frac{r}{R_e})
$$

$$
\frac{d^{2}y}{dt^2} = -{g_0}(\frac{r}{R_e}) + {\alpha} |\frac{dy}{dt}|^{\gamma}
$$

$$
r = R_e - y
$$

```python
# gravity

# (1) Velocity and position as function of time assuming g(y)

# Function
def dvdt(t, s):
    y, v = s
    dydt = v 
    dvdt = -g0*((R_e - np.abs(y))/R_e) - a*v**gamma # (R_e - y) is distance from center, with y and depth 
        # g(r) gets bigger when r is bigger
    return [dydt, dvdt]

# Get Event
def crash(t, s):
    return s[0]+4000

# Constants 
a = 0 # drag still off
gamma = 2

# Time span
t0, tf = 0, 30
t_eval = np.linspace(t0, tf, 10000)

# Initial conditions
y0 = [0, 0] # initial position (depth into shaft) and velocity

soln = solve_ivp(fun = dvdt, t_span=(t0, tf), y0=y0, t_eval = t_eval, events = crash)

print(soln)

# Plot
t = soln.t
y = soln.y[0]
v = soln.y[1]

fig, ax = plt.subplots()
ax.plot(t, y, color = "royalblue")
ax.set_ylabel("Position (m)", color = "royalblue")
ax.tick_params(axis="y", labelcolor="royalblue")

ax.set_xlabel("Time (s)")

ax2 = ax.twinx()
ax2.plot(t, v, color="deeppink", linestyle = ":")
ax2.set_ylabel("Velocity (m/s)", color = "deeppink")
ax2.tick_params(axis="y", labelcolor="deeppink")

#ax2.axhline(-50, color = "black", linestyle = "--", lw = 0.5)

ax.set_title("Height-Depedent g")

#fig.suptitle("Figure 2");

print(f"Crash time is now {soln.t_events[0]}")
```

### Height dependent g = g(r) = ${g_0}(\frac{r}{R_e})$ means gravity gets weaker as it goes closer, so compared to a constant g (using the force of gravity at the surface), the crash time is now slightly longer as the mass's average acceleration is lower, making its average speed lower.

### Finding ${\rho}_n$

$$
M_E = 4{\pi}\int_{0}^{R_E} {\rho}_n(1-\frac{r^2}{R_E^2})^n\,r^2 \,dr\
$$

$$
{\rho}_n = \frac{M_E}{4\pi \int_{0}^{R_E} (1-\frac{r^2}{R_E^2})^n\,r^2 \,dr\ }
$$

```python
# (2)
# quad(function, lower limit, upper limit)

def density(pn, r, n):
    return pn*((1 - (r/R_e)**2)**n)

def density_int(pn, r, n): # total function inside the integral 
    dens = pn*((1 - (r/R_e)**2)**n)
    return dens*(r**2)


# Function for array of interior masses given interior densities...

def M(r, pn, n):
    def density_int(r_prime, pn, n): # for dr', make sure r_prime is first arg
        dens = pn*((1 - (r_prime/R_e)**2)**n)
        return dens*(r_prime**2)
        
    M = np.zeros_like(r)
    for i, r_val in enumerate(r):
        M[i] = 4*np.pi * quad(density_int, 0, r_val, args = (pn, n))[0]
    return M


radii = np.linspace(0, R_e, 2000)


# Get correct rho
def inside_int(r, n): # total function inside the integral 
    dens = ((1 - (r/R_e)**2)**n)
    return dens*(r**2)

def rho(n):
    integral = quad(inside_int, 0, R_e, args = (n,))[0]
    rho_n = M_e / ( (4*np.pi)*(integral))
    return rho_n


M0 = M(radii, rho(0), 0)
M1 = M(radii, rho(1), 1)
M2 = M(radii, rho(2), 2)
M9 = M(radii, rho(9), 9)


fig, ax = plt.subplots(figsize = (7,6))
ax.plot(radii, M0, label = "n = 0")
ax.plot(radii, M1, label = "n = 1")
ax.plot(radii, M2, label = "n = 2")
ax.plot(radii, M9, label = "n = 9")
ax.set_ylabel("Mass in kg")
ax.set_xlabel("Radius in m")
ax.axhline(M_e, linestyle = ":", label = "Total Mass of the Earth")
ax.axvline(R_e, linestyle = "--", color = "black", linewidth = 1, label = "Radius of the Earth")
ax.legend();
ax.set_title("M(R)");
```

```python
def M(r, pn, n):
    def density_int(r_prime, pn, n): # for dr', make sure r_prime is first arg
        dens = pn*((1 - (r_prime/R_e)**2)**n)
        return dens*(r_prime**2)
        
    M = np.zeros_like(r)
    for i, r_val in enumerate(r):
        M[i] = 4*np.pi * quad(density_int, 0, r_val, args = (pn, n))[0]
    return M

def f_per_m(M, r):
    f = np.where(r==0, 0, G*M/(r**2))
    return f

f0 = f_per_m(M0, radii) # M0, radii are arrays
f1 = f_per_m(M1, radii)
f2 = f_per_m(M2, radii)
f9 = f_per_m(M9, radii)

fig2, ax2 = plt.subplots()
ax2.plot(radii, f0, label = "n = 0")
ax2.plot(radii, f1, label = "n = 1")
ax2.plot(radii, f2, label = "n = 2")
ax2.plot(radii, f9, label = "n = 9")
ax2.legend();

ax2.set_ylabel("F/m in N/kg")
ax2.set_xlabel("Radius in m")

ax2.set_title("Force Profiles for Different Powers n")
fig2.suptitle("Figure 6");


# Notes
    # n = 0 (constant density) looks good
    # n = 1, density prop to r^2, M(r) prop to r^5, F prop to r^3
    # n = 2, density prop to r^4, M(r) prop to r^7, F prop to r^5
    # n = 9, density prop to r^18, M(r) prop to r^21, F prop to r^19

n = [0, 1, 2, 9]
for i, n_val in enumerate(n):
    print(f"For n = {n_val}, pn = {rho(n_val)}")
```

```python
# (3)
# Position and velocity as a function of "time"
    # need the acceleration at any given point -- this is f = F/m

# Radii up to Earth, but R0 = 1 m to avoid singularity when calculated Fg 
radii = np.linspace(-R_e, R_e, 2000)
'''
Note that we are calculating radii from center outward. 
'''

def M(r, pn, n):
    '''
    First, define the integrand inside the density integral.

    To use quad, we define r_prime as the variable of integration to distinguish from r the actual radius later
    '''
    def density_int(r_prime, pn, n): # for dr', make sure r_prime is first arg
        dens = pn*((1 - (r_prime/R_e)**2)**n)
        return dens*(r_prime**2)

    '''
    Then, numerically calculate the integral using quad() up to a point "r"
        This is the mass M enclosed by radius "r", M(r)

    First we create an empty array with the same size as the array for "r" we take

    Then, for each radius value in r_val in array r, compute the mass enclosed M(r) numerically
        Since quad gives both sum and error, just take sum (index 0)

    Then, we add that value of M(r) to the array M, with the index of each r_val in array r
    matching with the index of the corresponding value in array M
    '''
    M = np.zeros_like(r)
    for i, r_val in enumerate(r): 
            # r_val gives access to entries inside array "r", enumerate pairs each r_val with its index
        M[i] = 4*np.pi * quad(density_int, 0, r_val, args = (pn, n))[0]
    return M

def f_per_m(M, r):
    '''
    M and r are both arrays, but they are "linked" with matching indices (same size)
    '''
    f = np.where(r==0, 0, G*M/(r**2))
    return f

'''
Force is always calculated with respect to "R"
and radii in this function only goes from 1 to R_e
    --> to go to other side, 1 to 2*R_e (?)
'''

# Get arrays for enclosed masses, for each density configuration
M0 = M(radii, rho(0), 0) 
M1 = M(radii, rho(1), 1)
M2 = M(radii, rho(2), 2)
M9 = M(radii, rho(9), 9)

# Get the force (per unit mass) of gravity of each mass array, as function of distance 
f0 = f_per_m(M0, radii)
f1 = f_per_m(M1, radii)
f2 = f_per_m(M2, radii)
f9 = f_per_m(M9, radii)

# Make the arrays for force into something we can use, a "function"
    # Using interpolation to make it useable; too hard to match with the force as an array 
f0_funct = interp1d(radii, f0, bounds_error= False) 
f1_funct = CubicSpline(radii, f1)
f2_funct = CubicSpline(radii, f2)
f9_funct = interp1d(radii, f9, bounds_error= False)

# Radii and y:
    # radii starts from center of Earth
    # y starts from Surface
    # y = R_e - r
    # r = R_e - y

def dvdt(t, s, f):
    y, v = s
    dydt = v
    dvdt = -f(R_e - np.abs(y))
        # force per m at the particular y value
        # r = R_e - y
    return [dydt, dvdt]

# Time span
t0, tf = 0, 5400
t_eval = np.linspace(t0, tf, 1000)

# Initial conditions
y0 = [0, 0] # start at y = 0 surface


soln_0 = solve_ivp(fun = dvdt, 
                   t_span=(t0, tf), 
                   y0=y0, 
                   t_eval = t_eval, 
                   args = (f0_funct,))

soln_1 = solve_ivp(fun = dvdt, 
                   t_span=(t0, tf), 
                   y0=y0, 
                   t_eval = t_eval, 
                   args = (f1_funct,))

soln_2 = solve_ivp(fun = dvdt, 
                   t_span=(t0, tf), 
                   y0=y0, 
                   t_eval = t_eval, 
                   args = (f2_funct,))

soln_9 = solve_ivp(fun = dvdt, 
                   t_span=(t0, tf), 
                   y0=y0, 
                   t_eval = t_eval, 
                   args = (f9_funct,))


# Plot
t0, t1, t2, t9 = soln_0.t, soln_1.t, soln_2.t, soln_9.t
y0, y1, y2, y9 = soln_0.y[0], soln_1.y[0], soln_2.y[0], soln_9.y[0]
v0, v1, v2, v9 = soln_0.y[1], soln_1.y[1], soln_2.y[1], soln_9.y[1]


fig, (axy, axv) = plt.subplots(2,1, figsize = (8,12))
divisor = 1000
axy.plot(t0, y0/divisor, label = "n = 0")
axy.plot(t1, y1/divisor, label = "n = 1")
axy.plot(t2, y2/divisor, label = "n = 2")
axy.plot(t9, y9/divisor, label = "n = 9")

axy.axhline(-12756, label = "Full Diameter of Earth", linewidth = 0.7, linestyle = ":")
axy.axhline(-12756/2, label = "Center", linewidth = 0.7, linestyle = ":", color = "black")


axy.set_title("Position vs Time")
axy.set_ylabel("Position from starting point (km)")
axy.set_xlabel("Time (s)")
axy.legend()

axv.plot(t0, v0, label = "n = 0")
axv.plot(t1, v1, label = "n = 1")
axv.plot(t2, v2, label = "n = 2")
axv.plot(t9, v9, label = "n = 9")

axv.set_title("Velocity vs Time")
axv.set_ylabel("Velocity (m/s)")
axv.set_xlabel("Time (s)")
axv.legend()

fig.suptitle("Figure 7")
fig.tight_layout();
```

### Relating orbital time (and thus fall time) with density:

$$
T_{\text{orb}} = \sqrt{\frac{4{\pi}^2R^3}{GM(R)}}
$$

$$
M(R) = \frac{4}{3}{\pi}R^3 \rho_n
$$

$$
T_{\text{orb}} = \sqrt{\frac{4{\pi}^2R^3}{\frac{4}{3}{\pi}R^3 \rho_n}}
$$

$$
T_{\text{orb}} \propto \frac{1}{\sqrt{\rho_n}}
$$

On a broad level, this makes sense conceptually, as greater density means greater mass, which means stronger gravity, which correlates to faster fall time. Density $\propto$ Mass $\propto$ $F_g$ $\propto$ acceleration, which means faster speed.
